**Function for fetching data from neo4j database**

```python
def run_query(query, parameters=None):
    with driver.session() as session:
        results = session.run(query, parameters)
        return [dict(record) for record in results]
```

Fetching Player Data and Cleaning it to change to a List

```python
query = """
        MATCH (p:PLAYER)
        RETURN p
        """
result = run_query(query)
# first we need to change the list of neo4j graph node to a list of dictionary
dummy_ls = []

for i in result:
    dictionary_value = dict(i.get("p"))
    dummy_ls.append(dictionary_value)

print(dummy_ls)

# I need to change the dictionary in to paragraph that is makes it easy for the embedding
paragraph = """ """
All_players = []
for i in dummy_ls:
  string = ""
  player = i.get("name")
  name = f" {player} is a player."
  paragraph += name
  string += name
  player_age = i.get("age")
  age = f" {player} is {player_age} years old."
  paragraph += age
  string += age
  player_weight = i.get("weight")
  weight = f" {player} is weights {player_weight} kilograms."
  paragraph += weight
  string += weight
  player_number = i.get("number")
  number = f" {player} wears {player_number} number Tshirt."
  paragraph += number
  string += number
  player_height = i.get("height")
  height = f" {player}'s height is {player_height}."
  paragraph += height
  string += number
  All_players.append(string)

print(All_players)
```

Fetching Players relation with eachother and cleaning it to change to List

```python
query = """MATCH (n: PLAYER)-[r]->(m: PLAYER)
RETURN n, r, m"""
new_result = run_query(query)
PL_PL_relations = []
for i in new_result:
  single_value = dict(i).get("r")
  first_node = dict(single_value.start_node).get("name")
  last_node = dict(single_value.end_node).get("name")
  relation = single_value.type
  string = f"{first_node} and {last_node} are {relation}."
  PL_PL_relations.append(string)
for j in PL_PL_relations:
  print(j)
```

Get the data from database using the query

```python
def get_data_from_db(query):
    # embed query
    xq = embed_model.get_text_embedding(query)
    # query database
    result = pinecone_index.query(vector=xq, top_k=20, includeMetadata=True)
    matches = []
    for i in result['matches']:
      # filter by score
      if (i['score'] > 0.6):
        # append best results
        matches.append(i.metadata['text'])
    return str(matches)
```

Getting Prompt

```python
B_INST, E_INST = "[INST]", "[/INST]"
B_SYS, E_SYS = "<<SYS>>\n", "\n<</SYS>>\n\n"
DEFAULT_SYSTEM_PROMPT = """\
  You will be given texts related to a certain topic. Write a summary on response that answers the question based on what is discussed in the texts in a single sentense.
  Do not mention anything outside of what is provided. Don't answer anything outside the context you are provided.
  If there isn't enough context, simply reply "This topic was not discussed previously"
  """

SYSTEM_PROMPT = B_SYS + DEFAULT_SYSTEM_PROMPT + E_SYS

def get_prompt(instruction):
    prompt_template =  B_INST + SYSTEM_PROMPT + instruction + E_INST
    return prompt_template
```

Format Context alongwith query

```python
def format_prompt(query, context):
    return '''
    ### Texts:
    {context}

    ### Question:
    {query}
    '''.format(context=context, query=query)
```
